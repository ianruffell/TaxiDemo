```python
from pygridgain import Client
import pandas as pd
import matplotlib.pyplot as plt

client = Client()
query = 'SELECT PULOCATIONID, count(*) FROM TRIPPICKUP GROUP By PULOCATIONID'
client.connect('127.0.0.1', 10800)
```

```python
locs = []
cars = []
with client.sql(query) as cursor:
    for row in cursor:
        locs.append(row[0])
        cars.append(row[1])

data = {'Location': locs, 'Cars' : cars}
```

```python
df = pd.DataFrame(data)
df.plot(kind='scatter', x='Location', y='Cars')
```
